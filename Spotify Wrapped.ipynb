```python
import requests
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import time
import datetime
```

```python
ClientId = "YOUR-CLIENT-ID"
ClientSecret = "YOUR-CLIENT-SECRET"
RedirectURI = "http://mywrapped.com"   ## you can set whatever you want
Scope = "user-top-read"                # mandatory
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=ClientId,
                                              client_secret=ClientSecret,
                                              redirect_uri=RedirectURI,
                                              scope=Scope))
```

```python
top_tracks_long = sp.current_user_top_tracks(limit=40,offset=0,time_range="long_term")
top_tracks_long

## shor_term(4weeks), medium_term(6months), long_term(alltime)
```

```python
def get_track_ids(top_tracks):
    track_ids = []
    for song in top_tracks['items']:
        track_ids.append(song['id'])
    return track_ids
```

```python
track_ids = get_track_ids(top_tracks_long)
track_ids
```

```python
new = sp.track('65ShmiE5aLBdcIGr7tHX35')
print(new)
```

```python
def get_track_info(id):
    data = sp.track(id)
    name=data['name']
    album = data['album']['name']
    artist = data['album']['artists'][0]['name']
    spotify_url=data['external_urls']['spotify']
    track_info = [name,album,artist,spotify_url]
    return track_info
```

```python
get_track_info('65ShmiE5aLBdcIGr7tHX35')
```

```python
tracks =[]
for i in range(len(track_ids)):
    time.sleep(1)
    track = get_track_info(track_ids[i])
    tracks.append(track)
```

```python
tracks
```

```python
df = pd.DataFrame(tracks,columns = ['name','album','artist','spotify_url'])
df
```

```python
df.to_csv('spotify_wrapped.csv',header=True,index=False)
```
